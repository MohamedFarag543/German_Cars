```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# These are the numerical attributes of our cars dataset
# Note that the "price" variable is the target variable and is not part of this list
num_attribs = ['mileage','hp','age']

# This is a pipeline that we will apply to the numerical attributes
# The purpose of this pipeline is to rescale the numerical attributes
# to be between 0 and 1
num_pipeline = Pipeline([
    ('min_max_scaler', MinMaxScaler()),
])

# These are the categorical attributes of our dataset
cat_attribs = ['make','model','fuel','gear']

# Define a ColumnTransformer object which will transform the columns of our dataset 
# by applying the numerical pipeline (min-max scaler) to each numerical attribute
# and applying the one-hot encoder to each categorical attribute
full_pipeline_cars = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs),
])

# Fit the full pipeline to the cars dataset
cars_prepared = full_pipeline_cars.fit_transform(cars)

# The target column is the 'price' 
labels_cars = cars['price']

# Define a linear regressor
lin_reg_cars = LinearRegression()

# Fit the linear regressor to the prepared cars dataset
lin_reg_cars.fit(cars_prepared, labels_cars)

# Use the resulting linear regression parameters to predict the price of each datapoint (car) in the dataset
price_predictions_cars = lin_reg_cars.predict(cars_prepared)

# Find the mean square error of our predictions over the whole dataset
lin_mse_cars = mean_squared_error(labels_cars, price_predictions_cars)

# Find the root mean square error
lin_rmse_cars = np.sqrt(lin_mse_cars)
print("RMSE of the linear regression price predictions for the cars data set: %.0f" % lin_rmse_cars)

# Find the mean absolute error
lin_mae_cars = mean_absolute_error(labels_cars, price_predictions_cars)
print('MAE of the linear regression price predictions for the cars data set: %.0f' % lin_mae_cars)

# Find the ratio of the RMSE to the mean price of cars in the dataset
print('Ratio of the RMSE to the mean price in the cars dataset: ' , lin_rmse_cars/(cars.price.mean()))

# Find the ratio of the MAE to the mean price of cars in the dataset
print('Ratio of the MAE to the mean price in the cars dataset: ' , lin_mae_cars/(cars.price.mean()))

# Compute the coefficient of determination
print('Coefficient of determination (R^2): ', lin_reg_cars.score(cars_prepared, cars['price']))
```
